```python
import numpy as np
import pandas as pd
import folium
```

```python
import warnings #경고 제어 라이브러리로 경고 메시지를 무시
#경고 메세지르 활성화 하고 싶을때 : (action='default')
warnings.filterwarnings(action='ignore') 
```

```python
m = folium.Map(
    location=[33.38077577868573, 126.53350416938764]
)

folium.Marker(
  location=[33.38077577868573,126.53350416938764],
  popup='한라산',
  icon=folium.Icon(color='red',icon='star')
).add_to(m)

m.save('map.html')
```

```python
m = folium.Map(
    location=[33.38077577868573, 126.53350416938764],
    #tiles = 'Stamen Terrain',  #타일 종류인데 우리는 사용하지 않습니다.
    zoom_start = 10
)
tooltip = 'Click!'

folium.Marker(
    [33.38077577868573, 126.53350416938764],
    popup ='한라산',
    tooltip = tooltip
).add_to(m)
m.save('index.html')
```

```python
m = folium.Map(
    location=[33.38077577868573, 126.53350416938764],
    #tiles = 'Stamen Terrain',  #타일 종류인데 사용하지 않겠다.
    zoom_start = 10
)

tooltip = 'Click!'


folium.Marker(
    [33.38077577868573, 126.53350416938764],
    popup ='한라산',
    tooltip = tooltip,
    icon = folium.Icon(color='#ff0000', icon= 'info-sign')
).add_to(m)

folium.Marker(
    [33.51210664966642, 126.49063423580104],
    popup ='제주국제공항',
    tooltip = tooltip
).add_to(m)

folium.Marker(
    [33.45801068752096, 126.56157022144995],
    popup ='제주대학교',
    tooltip = tooltip
).add_to(m)

m
```

```python
m = folium.Map(
    location=[33.38077577868573, 126.53350416938764],
    #tiles = 'Stamen Terrain',  #타일 종류인데 우리는 사용하지 않습니다.
    zoom_start = 10
)
tooltip = 'Click!'
for i in range(오름.shape[0]):
    print(오름.iloc[i]['오름명'], 오름.iloc[i]['위도'], 오름.iloc[i]['경도'])
    folium.Marker(
        [오름.iloc[i]['위도'],오름.iloc[i]['경도']],
        popup = 오름.iloc[i]['오름명'],
        tooltip = tooltip
    ).add_to(m)

m.save('map.html')
```

```python
m = folium.Map(
    location=[33.38077577868573, 126.53350416938764],
    #tiles = 'Stamen Terrain',  #타일 종류인데 우리는 사용하지 않습니다.
    zoom_start = 10
)
tooltip = 'Click!'
for i in range(오름.shape[0]):
    print(오름.iloc[i]['오름명'], 오름.iloc[i]['위도'], 오름.iloc[i]['경도'])
    folium.Marker(
        [오름.iloc[i]['위도'],오름.iloc[i]['경도']],
        popup = f'<div style="width:100px"><strong>{오름.iloc[i]["오름명"]}</strong><br>\
        주차장 : {오름.iloc[i]["주차장"]}<br> \
        화장실 : {오름.iloc[i]["화장실"]}<br> \
        <br></div>',
        tooltip = tooltip
    ).add_to(m)

m
```
